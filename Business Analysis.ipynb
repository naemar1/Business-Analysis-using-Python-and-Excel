```python
#Q4. Find the total sales generated by each State and display the results in descending order.

df.groupby("State")["TotalAmount"].sum()
```

```python
df.groupby("State")["TotalAmount"].sum().sort_values(ascending=False)
```

```python
#Q5. Find the total sales generated by each Category.
df.groupby("Category")["TotalAmount"].sum()
```

```python
#Customer & Product Analysis
```

```python
#Top 5 customers
#Q9. Display the Top 5 Customers based on total sales.
df.groupby("CustomerID")["TotalAmount"].sum().sort_values(ascending=False).head()
```

```python
#Top 7 customers
df.groupby("CustomerID")["TotalAmount"].sum().sort_values(ascending=False).head(7)
```

```python
#Payment Analysis
#Q11. Find the number of orders placed using each Payment Mode.
df.groupby("PaymentMethod")["OrderID"].count()
```

```python
#Q10. Find the total quantity sold for each product.

df.groupby("Quantity")["TotalAmount"].sum()
```

```python
df.groupby("Category")["TotalAmount"].sum().sort_values(ascending=False)
```

```python
#Time-based Analysis
#Q12. Calculate the monthly sales trend using the Order Date column.

df["Month"] = df["OrderDate"].dt.to_period("M")

df.groupby("Month")["TotalAmount"].sum()
```

```python
df.groupby("Month")["TotalAmount"].sum().sort_values(ascending=False)
```

```python
#Multi-level Analysis
#Q13. Find the sales generated by each State for every Category.

df.groupby(['Category','State'])['TotalAmount'].sum()
```

```python
#Visualization
```

```python
#Q15. Create a Bar Chart showing Sales by State.

df.groupby("State")["TotalAmount"].sum().plot(kind='bar')
plt.title('Revenue by State')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
df.groupby("State")["TotalAmount"].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Revenue by State')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
df.groupby(['CustomerID','CustomerName'])['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
#Q16. Create a Pie Chart showing Category-wise Sales.

df.groupby("Category")["TotalAmount"].sum().sort_values(ascending=False).plot(kind='pie')
plt.title('Revenue by Category')
plt.show()
```

```python
#Q17. Create a Line Chart showing Monthly Sales Trend.

df["Month"] = df["OrderDate"].dt.to_period("M")

df.groupby("Month")["TotalAmount"].sum().plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#Q18. Create a Horizontal Bar Chart showing the Top 10 Products by Sales.

df.groupby("ProductName")["TotalAmount"].sum().sort_values(ascending=False).head(10).plot(kind='barh')

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()
```
